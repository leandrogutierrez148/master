# 1 - Evaluación de la hipótesis 

La selección del modelo es el problema de elegir uno entre un conjunto de modelos candidatos.

Es común elegir un modelo que tenga el mejor rendimiento en un conjunto de datos de prueba de retención o estimar el rendimiento del modelo mediante una técnica de remuestreo, como *k-fold cross-validation.*

Un enfoque alternativo para la selección de modelos implica el uso de medidas estadísticas probabilísticas que intentan cuantificar tanto el rendimiento del modelo en el conjunto de datos de entrenamiento como la complejidad del modelo. Los ejemplos incluyen el criterio de información de **Akaike** y **BIC (Bayesian Information Criterion)**.

El beneficio de estas estadísticas de criterio de información es que no requieren un conjunto de pruebas de retención, aunque una limitación es que no tienen en cuenta la incertidumbre de los modelos y pueden terminar seleccionando modelos que son demasiado simples.

- La longitud mínima de descripción proporciona otro método de puntuación (criterio) de la teoría de la información que puede demostrarse que es equivalente al BIC.

Esto puede aplicarse en el aprendizaje no supervisado, para elegir un modelo de agrupamiento o aprendizaje supervisado, para escoger un modelo predictivo para una tarea de regresión o clasificación y, también, puede ser una subtarea de modelado, como la selección de características para un modelo dado.

Hay muchos enfoques comunes que pueden usarse para la selección de modelos. Por ejemplo, en el caso del aprendizaje supervisado, los tres enfoques más comunes son:

> - Conjuntos de datos de entrenamiento, validación y prueba.
> - Métodos de remuestreo.
> - Estadística probabilística.

En resumen, tenemos:

1. **El método más simple y confiable de selección de modelos** implica ajustar modelos candidatos en un conjunto de entrenamiento, ajustarlos en el conjunto de datos de validación y seleccionar un modelo que funcione mejor en el conjunto de datos de prueba de acuerdo con una métrica elegida, como precisión o error. **Un problema con este enfoque requiere muchos datos**.

2. **Las técnicas de remuestreo** intentan lograr lo mismo que el enfoque (train / val / test) para la selección del modelo, aunque utilizando un pequeño conjunto de datos. Un ejemplo es el k-fold cross-validation, Boosting o el mismo Bagging donde un conjunto de entrenamiento se divide en muchos pares de **entrenamientos / pruebas** y se ajusta y evalúa un modelo en cada uno. Esto se repite para cada modelo y se selecciona el que tenga la mejor puntuación promedio en los **k-pliegues/flogs**. Un problema con este y el enfoque anterior es que solo se evalúa el rendimiento del modelo, independientemente de la complejidad del modelo.

3. Un tercer enfoque para la selección de modelos es intentar **combinar la complejidad del modelo con su rendimiento en una puntuación,** para luego seleccionar el modelo que minimice o maximice la puntuación.

  Podemos referirnos a este enfoque como selección de modelo estadístico o probabilístico, ya que el método de puntuación utiliza un marco probabilístico.

## 2.1 - Selección de modelo probabilístico

La selección de **modelos probabilísticos** (*criterios de información*) proporciona una técnica analítica para calificar y elegir entre modelos candidatos.

> **Los modelos se puntúan según su rendimiento en el conjunto de datos de entrenamiento y según su complejidad**.

- **Rendimiento del modelo**: qué tan bien se ha desempeñado un modelo candidato en el conjunto de datos de entrenamiento.

- **Complejidad del modelo**: qué tan complicado es el modelo candidato entrenado después del entrenamiento.

El rendimiento del modelo puede evaluarse utilizando un marco probabilístico, como la verosimilitud logarítmica en el marco de la estimación de máxima verosimilitud. La complejidad del modelo puede evaluarse como el número de grados de libertad o parámetros del modelo.

> *Históricamente, se han propuesto varios "criterios de información" que intentan corregir el sesgo de máxima probabilidad mediante la adición de un término de penalización para compensar el ajuste excesivo de modelos más complejos.*

<img src="https://github.com/pifparfait/IL3/blob/main/Patter_recog.png?raw=true" width="400" height="600">

- Página 33, Reconocimiento de patrones y aprendizaje automático, 2006.

Una ventaja de los métodos de selección de modelos probabilísticos es que no se requiere un conjunto de datos de prueba, lo que significa que todos los datos se pueden usar para ajustar el modelo y el modelo final que se usará para la predicción en el dominio se puede calificar directamente.

Una limitación de los métodos de selección de modelos probabilísticos es que la misma estadística general no se puede calcular en una variedad de tipos diferentes de modelos. En cambio, la métrica debe derivarse cuidadosamente para cada modelo.

> *Cabe señalar que la estadística AIC está diseñada para comparaciones planificadas previamente entre modelos (a diferencia de las comparaciones de muchos modelos durante búsquedas automáticas)*.


<img src="https://github.com/pifparfait/IL3/blob/main/Applied_pred_mod.png?raw=true" width="400" height="600">

- Página 493, Modelado predictivo aplicado, 2013.


Otra limitación de estos métodos de selección es que no tienen en cuenta la incertidumbre del modelo.

> **Sin embargo, estos criterios no tienen en cuenta la incertidumbre de los parámetros del modelo y, en la práctica, tienden a favorecer modelos demasiado simples.** 

- Página 33, Reconocimiento de patrones y aprendizaje automático, 2006.

**Hay tres enfoques estadísticos para estimar qué tan bien un modelo dado se ajusta a un conjunto de datos y qué tan complejo es el modelo.** Y se puede demostrar que cada uno es equivalente o proporcional entre sí, aunque cada uno se derivó de un marco o campo de estudio diferente.

Los enfoques son:

1. Criterio de información de **Akaike (AIC)**: Derivado de probabilidad frecuentista.

2. Criterio de información **bayesiano (BIC)**: Derivado de probabilidad bayesiana.

3. Longitud mínima de descripción **(MDL)**: Derivado de la teoría de la información.

Cada información estadística puede calcularse utilizando la probabilidad logarítmica del modelo o de uno cualquiera (bajo estudio) y los datos observados. La probabilidad logarítmica proviene de la estimación de máxima verosimilitud (*Maximum Likelihood Estimation*) en inglés, que es una técnica para encontrar u optimizar los parámetros de un modelo en respuesta a un conjunto de datos de entrenamiento.

Normalmente en *Estimación de máxima verosimilitud*, se desea maximizar la probabilidad condicional de observar los datos ($X$) dada una distribución de probabilidad específica y sus parámetros ($\theta$), expresada formalmente como:

$$P (X; \theta)$$

Donde $X$ es, la distribución de probabilidad conjunta de todas las observaciones del dominio del problema de $1$ a $n$. Siendo $n$ el número total de las observaciones.

$$P (x1, x2, x3,…, xn; \theta)$$

La distribución de probabilidad conjunta puede reformularse como la multiplicación de la probabilidad condicional de observar cada ejemplo dados los parámetros de distribución. Multiplicar muchas probabilidades pequeñas juntas puede ser inestable; como tal, es común reformular este problema como la suma de la probabilidad condicional logarítmica natural.

$$\sum_{i}^{n} log (P (x_{i}; \theta))$$

Dado el uso frecuente del logaritmo en la función de verosimilitud, se denomina comúnmente **función logarítmica de verosimilitud.**

La función de probabilidad logarítmica para problemas comunes de modelado predictivo, incluye el error cuadrático medio para la regresión (por ejemplo: regresión lineal) y la pérdida logarítmica (entropía cruzada binaria) para la clasificación binaria (por ejemplo: regresión logística).

Examinaremos más de cerca cada una de las tres estadísticas, **AIC**, **BIC** y **MDL**, en las siguientes secciones.

## 2.2. Criterio de información de Akaike
El criterio de información de **Akaike**, o **AIC**, es un método para calificar y seleccionar un modelo.

Lleva el nombre del desarrollador del método, *Hirotugu Akaike* [1], y puede demostrarse que tiene una base en la teoría de la información y la inferencia basada en frecuentistas.

Al derivar de un marco frecuentista no puede interpretarse como una aproximación a la probabilidad marginal.

<img src="https://github.com/pifparfait/IL3/blob/main/ML_Prob_Pers.jpg?raw=true" width="400" height="600">

- Página 162, Aprendizaje automático: una perspectiva probabilística, 2012.


[1] H. Akaike, "A new look at the statistical model identification," in IEEE Transactions on Automatic Control, vol. 19, no. 6, pp. 716-723, December 1974, doi: 10.1109/TAC.1974.1100705.

La estadística AIC se define para la regresión logística de la siguiente manera (fuente de "Los elementos del aprendizaje estadístico").

<img src="https://github.com/pifparfait/IL3/blob/main/Elements_Stats.png?raw=true" width="400" height="600">

Existe varias formulas del AIC, pero usaremos la siguiente:

$$ \bbox[5px,border: 2px solid blue]{ 
AIC = \frac{-2}{N}* L + 2 * \frac{k}{N}
} $$


Donde:

* $N$ es el número de ejemplos en el conjunto de datos de entrenamiento

* $L$ es la probabilidad logarítmica del modelo en el conjunto de datos de entrenamiento y

* $k$ es el número de parámetros en el modelo.

>**NOTA**: Dado un conjunto de modelos candidatos para los datos, el modelo preferido es el que tiene el valor mínimo en el *AIC*. Por lo tanto *AIC* no solamente recompensa la bondad de ajuste, sino también incluye una penalidad, que es una función creciente del número de parámetros estimados. Esta penalización desalienta el sobreajuste (aumentando el número de parámetros libres en el modelo y mejora la bondad del ajuste, sin importar el número de parámetros libres en el proceso de generación de datos).

La puntuación, como se define anteriormente, se minimiza, por Ejp. se selecciona el modelo con el AIC más bajo.

Para utilizar AIC para la selección del modelo, simplemente elegimos el modelo que proporciona el AIC más pequeño sobre el conjunto de modelos considerados.
- Página 231, Los elementos del aprendizaje estadístico, 2016.

En comparación con el método **BIC** que se verá a continuación, la estadística **AIC** penaliza menos los modelos complejos, lo que significa que puede poner más énfasis en el rendimiento del modelo en el conjunto de datos de entrenamiento y, a su vez, seleccionar modelos más complejos.


<br>
<p> <mark>OBSERVACIONES</mark> </p>
<hr>

*Se ve que la penalización para **AIC** es menor que para **BIC**. Esto hace que AIC elija modelos más complejos*. 

- Página 162, Aprendizaje automático: una perspectiva probabilística, 2012.

## 2.3 - Bayesian Information Criterion (BIC)
El Criterio de Información Bayesiano, o **BIC**, es un método para calificar y seleccionar un modelo. Viene del campo de la probabilidad e inferencias Bayesianas. Al igual que AIC, es apropiado para modelos que se ajustan al marco de estimación de máxima verosimilitud (Maximum Likelihood Estimation).

La estadística **BIC** se calcula para la **regresión logística** de la siguiente manera (fuente de "Los elementos del aprendizaje estadístico"):

$$ \bbox[5px,border: 2px solid blue]{ 
BIC = -2 * L + log (N) * k
} $$


Donde:
* $log ()$ tiene la base-$e$ llamada logaritmo natural

* $L$ es la probabilidad logarítmica del modelo

* $N$ es el número de ejemplos en el conjunto de datos de entrenamiento y 

* $k$ es el número de parámetros en el modelo


<br>
<p> <mark>OBSERVACIONES</mark> </p>
<hr>

La puntuación como se define arriba se minimiza, seleccionando por ejemplo, el modelo con el **BIC** más bajo.

>**NOTA** La cantidad calculada es diferente de la **AIC**, aunque puede demostrarse que es proporcional a la **AIC**. A diferencia del **AIC**, el **BIC** penaliza más al modelo por su complejidad, lo que significa que los modelos más complejos tendrán una puntuación peor (mayor) y, a su vez, será menos probable que sean seleccionados.

> Ten en cuenta que, en comparación con **AIC**, esto penaliza más la complejidad del modelo. 
- Página 217, Reconocimiento de patrones y aprendizaje automático, 2006.

Es importante destacar que la derivación de **BIC** bajo el marco de probabilidad bayesiano significa que si una selección de modelos candidatos incluye un modelo verdadero para el conjunto de datos, entonces la probabilidad de que **BIC** seleccione el buen modelo aumenta con el tamaño del conjunto de datos de entrenamiento. Esto no se puede decir del criterio **AIC**.

> **NOTA** Dada una familia de modelos, incluido el modelo verdadero, la probabilidad de que BIC seleccione el modelo correcto se aproxima a uno como el tamaño de la muestra $N \to\infty$. 
- Página 235, Los elementos del aprendizaje estadístico, 2016.

<br>
<p> <mark>IMPORTANTE</mark> </p>
<hr>


*Una desventaja de **BIC** es que para conjuntos de datos de entrenamientos más pequeños y menos representativos, es más probable elegir modelos que sean demasiado simples.*

# 3 - Caso de uso en la Regresión lineal

El objetivo de este caso de uso es validar un modelo usando los criterios de información **AIC** y **BIC**.

En este caso, se usa un problema de regresión de prueba proporcionado por la función *scikit-learn make_regression ()*. El problema tendrá dos variables de entrada y requerirá la predicción de un valor numérico objetivo. Después se podrá aplicar esta metodologia en datos más complejos. 

Es preciso que para un uso academico, se sepa de dónde proviene la formulación de los criterios de información, teniendo en cuenta que, casi todas las librerias ya los tienen implementados. Así, que el trabajo del Data scientist sería saber usar dichos criterios en cada uno de los casos que se le presenta.


## 3.1 - Cálculo del criterio Akaike (AIC) para un modelo de regresión lineal

```python
#importamos las librerias necesarias
from math import log
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

 
# Cálculo de AIC para la regresión lineal
def calculate_aic(n, mse, num_params):
	aic = n * log(mse) + 2 * num_params
	return aic

# calculate bic for regression
def calculate_bic(n, mse, num_params):
    bic = n * log(mse) + num_params * log(n)
    return bic

# Generamos dataset
X, y = make_regression(n_samples=100, n_features=2, noise=0.1)

#definimos y ajustamos el modelo a todos los datos
model = LinearRegression()
model.fit(X, y)

# número de parametros del modelo. Un buen Analista, deberia conseguir un modelo funcional con el minimo
#número de parametros. El objetivo es buscar la eficiencia entre el coste computacional y el error aceptable
num_params = len(model.coef_) + 1
print('Number of parameters: %d' % (num_params))

#predecir el conjunto de entrenamiento
yhat = model.predict(X)

# calcular el error. Este paso es lo que nos determina si tenemos un buen modelo o no.
mse = mean_squared_error(y, yhat)
print('MSE: %.3f' % mse)

# Cálculo del criterio AIC. Con este criterio de información podremos saber o no, si nuestro modelo es bueno para el caso 
#bajo estudio.
aic = calculate_aic(len(y), mse, num_params)
print('AIC: %.3f' % aic)
```

```python
# Calculo del BIC. Con este criterio de información podremos saber o no, si nuestro modelo es bueno para el caso 
#bajo estudio.
bic = calculate_bic(len(y), mse, num_params)
print('BIC: %.3f' % bic)
```


**Estrategia forward**
1. Se construyen todos los modelos posibles con las variables $x_{1}, \dots, x_{n}$ y se selecciona la variable para el modelo con menor criterio de información. 

2. Con la variable seleccionada ya fija, se construyen modelos con todas las combinaciones de dos variables, es decir, ($x_{i}, x_{i-n}$) con $i$ diferente de $n$. Y nos quedamos con la pareja con el menor criterio de información.

3. Se combinan las variables ya seleccionadas, con una de las variables no seleccionadas, y se va al paso (2) nuevamente.

4. Se repite todo el proceso hasta que no se puedan adicionar más variables. El criterio de información obtenido para $n$ variables debe ser inferior que el obtenido para todos los modelos con $n+1$ variables.



**Estrategia backward** (divide y vencerás)
1. Se parte de un modelo con todas las variables seleccionadas por el analista.

2. Se consideran todos los modelos resultantes al eliminar una de las variables seleccionadas. Se elimina la variable que produce el modelo con el menor valor del criterio de información.

3. Se repite el proceso hasta que no quedan variables irrelevantes.
