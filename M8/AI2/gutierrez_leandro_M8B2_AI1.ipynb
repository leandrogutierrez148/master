# Enunciado
Antes de afrontar cualquier tipo de problema es importante analizarlo para aplicar la metodología o marco de trabajo adecuado en función a su complejidad. 

A continuación se describen diferentes escenarios: 

- **Escenario 1**: misión a la luna
Para permitir a futuras misiones de explotación de los recursos de la luna, la NASA quiere realizar una misión de exploración del satélite y cartografiar toda su superficie con imágenes de alta definición y alto detalle del subsuelo. Durante varios días la misión no tripulada orbitará alrededor de la luna registrando la información necesaria y volver 2 semanas después con lo recabado. La tecnología de exploración la proporciona un contratista privado.

- **Escenario 2**: un plato sorprendente a base de insectos
Un chef profesional quiere crear un nuevo plato para su restaurante con 2 estrellas Michelín donde ofrece productos locales con técnicas de vanguardia. Recientemente se ha interesado por los insectos, pero no tiene muy claro qué uso se puede hacer de ellos ni cómo cocinarlos o combinarlos con otros elementos. Lo que sí tiene claro es que quiere sorprender a sus clientes sin perder la esencia que le ha llevado al éxito.

- **Escenario 3**: una colonia permanente en Marte
Un empresario de gran éxito se ha propuesto establecer una colonia permanentemente habitada en Marte en esta década. Para ello ha fundado una empresa con la que construir naves espaciales y sus motores de tal manera que puedan salvar los más de siete meses que dura el viaje y trasladar toda la carga necesaria con un coste asumible. Aunque ha habido varios fallos catastróficos durante las pruebas de los motores, lo cierto es que no cesa en el intento y cada vez los cohetes parecen más fiables.

- **Escenario 4**: el sistema de navegación aérea se ha caído
El sistema de navegación aérea europeo ha dejado de funcionar sin ninguna explicación con cientos de vuelos en el aire. Ahora mismo no tenemos ninguna explicación de qué puede haber sucedido ni de las causas, pero la situación es crítica y hay muchas vidas en juego.

- **Escenario 5**: migración a una nueva tecnología
El cliente está muy contento con su aplicación de gestión de los gastos de sus empleados, pero tecnológicamente está anticuada. Por ello el cliente nos solicita la migración de su actual aplicación a dicha tecnología como expertos en la misma.

- **Escenario 6**: pandemia a la vista
A finales del año pasado se detectó un brote de una enfermedad respiratoria desconocida al otro lado del globo. Los primeros intentos para controlarla resultaron infructuosos y unos meses después aparecieron varios casos de contagio en nuestro país. A día de hoy se conoce algo más sobre el virus causante, su forma de transmisión y grado de mortalidad. Además, como en toda epidemia, las medidas de distanciamiento social son claves para frenar. El gobierno del país se enfrenta al dilema de acabar con el virus dañando lo menos posible la economía.

- **Escenario 7**: mejorar la toma de decisiones en los hospitales
Ante la crisis del COVID los hospitales necesitan tener información actualizada y fiable que les permita tomar decisiones con cierto nivel de confianza, por ejemplo para planificar la dedicación del personal, los espacios y otros recursos, la compra de material y fármacos, etc. Es posible que una aplicación con los indicadores adecuados les pudiese ayudar, pero tendríamos que explorar esta idea y validarla con los propios usuarios.

- **Escenario 8**: hoy comemos bien
Un cocinero amateur tiene que elaborar un plato siguiendo una receta de cocina molecular. Son más de 20 ingredientes y varias elaboraciones con esferificaciones, uso de nitrógeno, varios tipos de cocción, etc. La receta está descrita paso a paso y el cocinero ya ha practicado alguna de estas elaboraciones.

# Solución
En este trabajo práctico abordaremos los aspectos clave para la identificación y la gestión de la complejidad en diferentes escenarios hipotéticos. A partir de la utilización de la matriz de Stacey identificaremos la complejidad inherente a cada proyecto y justificaremos nuestra decisión. Además identificaremos riesgos y determinaremos la metodolgía de desarrollo adecuada.

En primer lugar es necesario visualizar la matriz de Stacey y ver de que trata:

![Opcion 1](/Users/lgutierrez/Proyectos/master/M8/AI2/images/stacey.png)

Vemos que en el eje horizontal tenemos el `COMO`, el cual nos indica que nivel de conocimientos o certidumbre que tenemos sobre el como debe resolverse el problema, esto implica conocimientos sobre el contexto donde se circunscribe el problema, conocimiento sobre las tecnologías intervinientes, etcetera. Mientras que en el eje vertical encontramos el `QUE`, y su recorrida nos indica que tanto o que tan poco se conoce sobre el problema que se desea resolver en sí.

Dentro de este marco podemos identificar diferentes cuadrantes o posiciones donde se pueden ubicar cada uno de los enunciados propuestos. Es así que, si por ejemplo tenemos alta certidumbre en el `QUE` y en el `COMO`, es decir, una posición cercana al origen de coordenadas podemos catalogar nuestro escenaria como **simple** o sencillo. De manera análoga si la certidumbre en el `QUE` y en el `COMO` es baja y nuestro escenaria se sitúara en el vértice superior derecho, podemos decir que nuestro problema de naturaleza **caotica**. En el medio de estos dos extremos encontramos matices, como ser las etiquetas **complicado**, para cuando los conocimientos son discretos, y **complejo** cuando aumentamos en incertidumbre.

Habiendo visto este pequeño resumen ahora procedemos a la clasificación de cada uno de nuestros caso de estudio:

#### Escenario 1: misión a la luna
- Matriz de Stacey
  - Grado de certidumbre **complicado**.
  - QUE: La NASA tiene claro el objetivo del proyecto, es decir la exploración y el estudio de la superficie de satélites con imágenes de alta definición. Los detalles de la necesidad están claros
  - COMO:La misión depende de la técnología de un contratista externo, por lo que los detalles de como se implementará tanto la nave como la máquina que obtenga las imágenes son desconocidas. Se puede esperar que los contratistas de la NASA tengan un alto grado de conocimiento sobre estas áreas mencionadas, por lo que se asume un valor conservador en cuanto a la incertidumbre en el COMO. 
- Riesgos
  - Riesgo: Posible fallas en la entrega o en los tiempos estipulados por parte del contratista externo.
  - Mitigación: Se pactan entregas periódicas sobre un calendario estipulado con un margen discreto para la puesta en órbita del equipamiento
- Estrategia: **Agile** para permitir iteraciones acumulativas y ajustes continuos.

#### Escenario 2: un plato sorprendente a base de insectos
- Matriz de Stacey
  - Grado de certidumbre **complicado**.
  - QUE: El chef tiene un grado de certidumbre bajo sobre lo que desea hacer, tiene nociones básicas de lo que espera lograr, pero no definiciones concretas de que plato presentar y como compartirlo con su clientela.
  - COMO: Como menciona el enunciado el chef no tiene muy claro qué uso se puede hacer de los insecto ni cómo cocinarlos o combinarlos con otros elementos. Sin embargo al ser un chef profesional conoce su cocina y su clientela. Tampoco sabe como se cocinan estos insectos.
- Riesgos
  - Riesgo: Posible rechazo por parte de sus clientes a estos platos que combinan insectos.
  - Mitigación: Pruebas de menú reducidas y con feedbacks directos.
- Estrategia: Agile para permitir iteraciones rápidas y con feedbacks de los clientes.

#### Escenario 3: una colonia permanente en Marte
- Matriz de Stacey
  - Grado de certidumbre **complejo**.
  - QUE: El conocimiento sobre el QUE está medianamente bien definido, los objetivos son claros, pero la incertidumbre sigue siendo alta dado que poco se conoce sobre el planeta Marte hasta el día de hoy. Hay inconmensurables situaciones que aún no se han presentado y que son de dificil antelación.
  - COMO: El COMO posee aún mas incertidumbre, la tecnología necesaria para establecer vida en Marte no está desarrollada. Tengamos en cuenta no es lo mismo llegar a Marte que establecer colonia, se deben definir y crear métodos que permitan la generación ininterrumpida de condiciones de vida, como ser el oxígeno, el agua y el alimento.
- Riesgos
  - Riesgo: Imposibilidad economica de llegar culminar el proyecto.
  - Mitigación: Prototipado evolutivo y busqueda de financiamiento externo contra pruebas de viabilidad.
- Estrategia: Agile para permitir iteraciones de prueba y error. 

#### Escenario 4: el sistema de navegación aérea se ha caído
- Matriz de Stacey
  - Grado de certidumbre **caótico**. 
  - QUE: El conocimiento sobre el QUE es nulo, no se sabe que sucedió.
  - COMO: El conocimiento sobre el COMO es también nulo, al desconocer la razón del problema se desconoce los elementos y tecnologías necesarias para su solución.
- Riesgos
  - Riesgo: Riesgo de muerte para miles de personas ante la caida del sistema principal.
  - Mitigación: Establecer protocolos de emergencia básicos para estas situaciones.
- Estrategia: Ninguno, se debe resolver el problema de manera inmediata.

#### Escenario 5: migración a una nueva tecnología
- Matriz de Stacey
  - Grado de certidumbre **simple**. 
  - QUE: El conocimiento del negocio y sus necesidades está claro por parte del cliente. Además tener una versión legacy de la aplicación ayuda para sentar las bases del nuevo desarrollo.
  - COMO: El como está claro, el equipo técnico es experto en la nueva tecnología requerida.
- Riesgos
  - Riesgo: Pérdida de datos.
  - Mitigación: Backups recurrentes y pruebas de prototipo en ambientes de desarrollo.
- Estrategia: Waterfall para asegurar la secuencia de desarrollo completa, desde la educción de requisitos, pasando por el diseño hasta el despliegue final en producción.

#### Escenario 6: pandemia a la vista
- Matriz de Stacey
  - Grado de certidumbre **complicado**. 
  - QUE: El grado de certidumbre sobre el QUE es escueto, no se conoce en plenitud el agente transmisor y es altamente complicado anticiparse a nuevos brotes, en general son cepas o variaciones nuevas de estos agentes.
  - COMO: Las herramientas con las que cuenta el estado son conocidas, y ya se tiene experiencia en otros casos de pandemias a gran escala como ser la del COVID-19. Pero siempre existe poca certidumbre de la respuesta de la economia ante este tipo de eventos.
- Riesgos
  - Riesgo: Colapso del sistema sanitario y daño a la economía
  - Mitigación: Desarrollo de modelos predictivos indicadores de posibles brotes en desarrollo.
- Estrategia: Agile para poder iterar y dar respuestas rápidas según el virus avance.

#### Escenario 7: mejorar la toma de decisiones en los hospitales
- Matriz de Stacey
  - Grado de certidumbre **complicado**. 
  - QUE: El conocimiento del requerimiento es claro y está bien definido.
  - COMO: Todavía no se conoce la tecnología que pueden dar la solución y se necesita validación por parte de los usuarios.
- Riesgos
  - Riesgo: Poca adaptación por parte de los usuarios.
  - Mitigación: Documentación clara y disponible. Validación a cada etapa.
- Estrategia: Agile nos permitiría hacer iteraciones pequeñas y tener feedbacks validadores a cada etapa.

#### Escenario 8: hoy comemos bien
- Matriz de Stacey
  - Grado de certidumbre **complicado**. 
  - QUE: El nivel de certidumbre sobre el QUE es alto, el requerimiento está claro.
  - COMO: El COMO a pesar de estar bien detallado y explicado en la receta, quien interpreta es un cocinero amateur por lo que se podría enfrentar a ciertas situaciones desconocidas. A medida que el cocinero avance en sus pruebas ganará la experiencia necesaria para disminuir esta incertidumbre.    
- Riesgos
  - Riesgo: Error en la ejecución de la receta.
  - Mitigación: Practica previa.
- Estrategia: Waterfall, el procedimiento es secuencial y está bien definido.
