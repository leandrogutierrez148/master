# Enunciado
El proceso y el trabajo emergentes deben ser visibles para aquellos que realizan el trabajo, así como para los que reciben el trabajo. Con Scrum, las decisiones importantes se basan en el estado percibido de sus tres artefactos formales. Los artefactos que tienen poca transparencia pueden conducir a decisiones que disminuyen el valor y aumentan el riesgo.

La transparencia permite la inspección. La inspección sin transparencia genera engaños y desperdicios. 

Rellena la siguiente matriz indicando para cada elemento de Scrum qué hacemos transparente y a quién es responsable de dicho elemento.

# Solución

| **Elemento de Scrum**      | **¿Qué hacemos transparente?**                                        | **¿De quién es la responsabilidad?**        |
|-----------------------------|------------------------------------------------------------------------|---------------------------------------------|
| **Product Backlog**         | La lista de requisitos en forma de lista ordenada y priorizada según el valor que aporta al proyecto.    | Product Owner                               |
| **Product Goal**            | Define el objetivo al que se quiere llegar con el trabajo sobre el Product Backlog. | Product Owner                               |
| **Sprint Backlog**          | Los elementos del Product Backlog seleccionados para el Sprint. | Product Owner, Scrum Master y Developers                                  |
| **Sprint Goal**             | El propósito específico del Sprint, definido por todo el Scrum Team.      | Product Owner, Scrum Master y Developers                                  |
| **Increment**               | La suma de los elementos terminados en el Sprint que cumplen con la Definition of Done. | Product Owner, Scrum Master y Developers                                  |
| **Definition of Done**      | Acuerdo que permite al equipo definir cuando un desarrollo está finalizado. | Product Owner, Scrum Master y Developers |
