```python
from datetime import datetime, timedelta

def proximo_miercoles(fecha_referencia=None):
    if fecha_referencia is None:
        fecha_referencia = datetime.today()

    dias_hasta_miercoles = (2 - fecha_referencia.weekday() + 7) % 7
    dias_hasta_miercoles = 7 if dias_hasta_miercoles == 0 else dias_hasta_miercoles

    return fecha_referencia + timedelta(days=dias_hasta_miercoles + 7)

prox_mierc = proximo_miercoles() 
```

```python
# <td class=" " data-handler="selectDay" data-event="click" data-month="6" data-year="2025"><a class="ui-state-default" href="#">23</a></td>
try:
    selector_fecha = driver.find_element(By.ID, "fechaTabla")
    selector_fecha.click()
    xpath = "//*[@data-handler='selectDay' and @data-month='6' and @data-year='2025' and a[text()='" + str(prox_mierc.day) +"']]"
    dia = driver.find_element(By.XPATH, xpath)
    dia.click()

except NoSuchElementException as e:
    print(f"No se encontró elemento 'selector de fechas'")
```

```python
try:
    hora = driver.find_element(By.XPATH, "//*[@columna='3' and @datahora='20:00']")
    hora.click()

except NoSuchElementException as e:
    print(f"No se encontró elemento 'celda de horario'")
```
